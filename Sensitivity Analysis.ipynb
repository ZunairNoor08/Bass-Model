```python
5#importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# defining the Bass Model Formula
def diffusion_process(t, p, q, m):
    # Calculate F(t)
    exp_term = np.exp(-(p + q) * t)
    F_t = m * (1 - exp_term) / (1 + (q * exp_term / p))

    # Calculate f(t)
    f_t = (p + (q / m) * F_t) * (m - F_t)
    return f_t, F_t

def main():
    # User inputs
    p_values = [float(input("Enter the first value of p: ")),
                float(input("Enter the second value of p: ")),
                float(input("Enter the third value of p: "))]
    q = float(input("Enter the value of q: "))
    m = float(input("Enter the value of m: "))
    t_start = float(input("Enter the start of the time range (t): "))
    t_end = float(input("Enter the end of the time range (t): "))
    num_points = int(input("Enter the number of points to plot: "))
    print(' ')

    # Time range in Years
    t = np.linspace(t_start, t_end, num_points)

    for p in p_values:
        f_t, F_t = diffusion_process(t, p, q, m)
        plt.plot(t, F_t, label=f'F(t) for p = {p}')
        plt.title('Cumulative Sales (S Curve)', fontsize=14)
        plt.xlabel('Year', fontsize=12)
        plt.ylabel('Cumulative Sales', fontsize=12)
        # Plotting for 2008 onwards
        plt.xticks(np.arange(t_start, t_end + 1, 2), labels=[f"{8 + int(x):02}" for x in np.arange(t_start, t_end + 1, 2)])
        plt.grid(True)
        plt.legend(fontsize=12)
        plt.show()
        print(' ')

    # Calculate f(t) and F(t)
    f_t, F_t = diffusion_process(t, p, q, m)

if __name__ == "__main__":
    main()
```
