**Bass Model - Germany**

```python
#importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# defining the Bass Model Formula
def diffusion_process(t, p, q, m):
    # Calculate F(t)
    exp_term = np.exp(-(p + q) * t)
    F_t = m * (1 - exp_term) / (1 + (q * exp_term / p))

    # Calculate f(t)
    f_t = (p + (q / m) * F_t) * (m - F_t)
    return f_t, F_t

def main():
    # User inputs
    p = float(input("Enter the value of p: "))
    q = float(input("Enter the value of q: "))
    m = float(input("Enter the value of m: "))
    t_start = float(input("Enter the start of the time range (t): "))
    t_end = float(input("Enter the end of the time range (t): "))
    num_points = int(input("Enter the number of points to plot: "))
    print(' ')

    # Time range in Years
    t = np.linspace(t_start, t_end, num_points)

    # Calculate f(t) and F(t)
    f_t, F_t = diffusion_process(t, p, q, m)

    # Plotting Rogers Bell Curve (Yearly Sales)
    plt.figure(figsize=(12, 6))
    plt.plot(t, f_t, label='Yearly Sales (f(t))', color='blue')
    plt.title('Yearly Sales (Bell Curve)', fontsize=14)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Yearly Sales', fontsize=12)
    plt.xticks(np.arange(t_start, t_end + 1, 2), labels=[f"{8 + int(x):02}" for x in np.arange(t_start, t_end + 1, 2)])
    plt.grid(True)
    plt.legend(fontsize=12)
    plt.show()
    print(' ')

    # Plotting S Curve (Cumulative Sales)
    plt.figure(figsize=(12, 6))
    plt.plot(t, F_t, label='Cumulative Sales (F(t))', color='green')
    plt.title('Cumulative Sales (S Curve)', fontsize=14)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Cumulative Sales', fontsize=12)
    plt.xticks(np.arange(t_start, t_end + 1, 2), labels=[f"{8 + int(x):02}" for x in np.arange(t_start, t_end + 1, 2)])
    plt.grid(True)
    plt.legend(fontsize=12)
    plt.show()
    print(' ')

    # Inputs for actual sales data
    actual_t_start = float(input("Enter the start of the time range for actual sales data (t): "))
    actual_t_end = float(input("Enter the end of the time range for actual sales data (t): "))
    actual_t = np.arange(actual_t_start, actual_t_end + 1, 1)
    print(f"Enter the actual yearly sales values for each year from {actual_t_start} to {actual_t_end}:")
    actual_sales = [float(input(f"Year {int(8 + year):02}: ")) for year in actual_t]
    print(' ')

    # Plotting Bell Curve with Actual Sales
    plt.figure(figsize=(12, 6))
    plt.plot(t, f_t, label='Modeled Yearly Sales (f(t))', color='blue')
    plt.plot(actual_t, actual_sales, label='Actual Yearly Sales', color='red', linestyle='-')
    plt.title('Yearly Sales: Modeled vs Actual', fontsize=14)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Yearly Sales', fontsize=12)
    plt.xticks(np.arange(t_start, t_end + 1, 2), labels=[f"{8 + int(x):02}" for x in np.arange(t_start, t_end + 1, 2)])
    plt.grid(True)
    plt.legend(fontsize=12)
    plt.show()

if __name__ == "__main__":
    main()
```
